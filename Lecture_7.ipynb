## Task

```python
a1 = np.array([70, 74, 74.5, 79, 85, 93, 94, 98, 106.5, 107])
a2 = np.array([64, 76.5, 67, 73.5, 89, 85, 89.5, 91, 98, 100.5])

a2 - a1
```

## Task
Задача: Даны заработные платы людей, принадлежащих к трем разным профессиям (условия нормальности не соблюдается).		gr 1: 70, 50, 64, 61, 75, 67, 73		gr 2: 80, 78, 90, 68, 74, 65, 85		gr 3: 141, 142, 140, 152, 161, 163, 155 Требуется определить, влияет ли профессия на заработную плату.

```python
gr_1 = np.array([70, 50, 64, 61, 75, 67, 73])
gr_2 = np.array([80, 78, 90, 68, 74, 65, 85])
gr_3 = np.array([141, 142, 140, 152, 161, 163, 155])

stats.kruskal(gr_1, gr_2, gr_3)
```
