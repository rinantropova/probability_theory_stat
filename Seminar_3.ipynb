```python
def quantile(n, k):
    q = (n * k) / 100
    if isinstance(q, float):
        j = int(q) + 1
    else:
        j = q
    return j
        
```

```python
q_75 = quantile(10, 75)
q3 = array[q_75-1]
```

```python
q_25 = quantile(10, 25)
q1 = array[q_25-1]
```

```python
iqr = q3 - q1
iqr
```
